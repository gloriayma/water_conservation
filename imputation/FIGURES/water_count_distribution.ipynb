# Solvent Count Distribution

Distribution of the number of solvent (water) molecules per structure in the filtered PDB.

**Filters applied (all from manifest.json — no npz reads):**
- Deposit date before 2023-06-01
- Method: X-ray diffraction only
- Resolution ≤ 9.0 Å
- Non-water residues ≤ 1024

**Notes on manifest fields:**
- `structure.deposited` — deposit date string (ISO format)
- `structure.method` — experimental method
- `structure.resolution` — resolution in Å
- `chains[i].mol_type == 4` — water chain; `num_residues` = water count
- Non-water residues = sum of `num_residues` for chains with `mol_type != 4`

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path(".").resolve()))

from water_dist_helpers import (
    load_manifest,
    filter_and_count,
    make_water_count_figure,
    MANIFEST_PATH,
)
```

## 1. Load manifest and count total structures

```python
manifest = load_manifest()
print(f"Total entries in manifest:  {len(manifest):,}")
print(f"(= number of .npz files in structures/)")
```

## 2. Apply filters and count waters

```python
filtered, stats = filter_and_count(manifest)

print("Filter pipeline:")
print(f"  Total in manifest:               {stats['total_in_manifest']:>8,}")
print(f"  After deposit date < 2023-06-01: {stats['after_deposit_date_filter']:>8,}")
print(f"  After method = x-ray:            {stats['after_method_filter']:>8,}")
print(f"  After resolution <= 9.0 Å:       {stats['after_resolution_filter']:>8,}")
print(f"  After residues <= 1024:          {stats['after_residues_filter']:>8,}")
print()
print(f"  Of those, with >= 10 waters:     {stats['n_geq10_waters']:>8,}")
```

## 3. Plot distribution

```python
import matplotlib.pyplot as plt

save_path = Path(".") / "water_count_distribution.pdf"

fig, ax = make_water_count_figure(filtered, stats, save_path=save_path)
plt.show()
```
